### Automated Accounting Procedure

```python
# import the needed packages
import pandas as pd
import numpy as np


### Function to calculate Total Sales for the Day
def daily_sales():
    try:
        total_morning_sales = float(input(f"Enter total morning sales: "))
        total_evening_sales = float(input(f"Enter total evening sales: "))
        daily_total_sales = total_morning_sales + total_evening_sales
        print(f'Total Sales for the day is {daily_total_sales} Naira.')
    except ValueError:      #error handling to ensure only floats and integers are accepted as inputs
        print('Invalid input. Please enter a valid number')
    
    
    
### Function to calculate Worker's Salary
def worker_salary(): 
    try:
        hourly_rate = float(input(f"Enter the hourly rate: "))
        hours_worked = float(input(f"Enter the hours worked: "))
        salary = hourly_rate * hours_worked
        print(f"The worker's salary is '{salary}' Naira.")
    except ValueError:        #error handling to ensure only floats and integers are accepted as inputs
        print('Invalid input. Please enter a valid number')
        
        
        
### Function to calculate Profit
def profit():
    try:
        total_sales = float(input(f"Enter total sales: "))
        total_cost = float(input(f"Enter total costs: "))
        profit = total_sales - total_cost
        print(f'Profit(or loss) is {profit} Naira.')
    except ValueError:       #error handling to ensure only floats and integers are accepted as inputs
        print('Invalid input. Please enter a valid number')
    
    
    
### Function to calculate Tips for shift
def shift_tips():
    try:
        total_morning_sales = float(input(f"Enter total sales for morning shift: "))
        total_evening_sales = float(input(f"Enter total sales for evening shift: "))
        morning_tips = total_morning_sales * 0.02
        evening_tips = total_evening_sales * 0.02
        print(f'Morning tip is {morning_tips} Naira.')
        print(f'Evening tip is {evening_tips} Naira.')
    except ValueError:       #error handling to ensure only floats and integers are accepted as inputs
        print('Invalid input. Please enter a valid number')  
   

### Function to calculate Total Tips for the Day
def daily_tips(): 
    try:
        total_morning_sales = float(input(f"Enter total sales for morning shift: "))
        total_evening_sales = float(input(f"Enter total sales for evening shift: "))
        morning_tips = total_morning_sales * 0.02
        evening_tips = total_evening_sales * 0.02
        daily_tips = morning_tips + evening_tips
        print(f'Total tips for the day is {daily_tips} Naira.')
    except ValueError:       #error handling to ensure only floats and integers are accepted as inputs
        print('Invalid input. Please enter a valid number')     


        
        
# User interface for the Main program - accounting process
def run_accounting():
    while True:
        print("\nAccounting System")
        print("1. Calculate Total Sales for the Day")
        print("2. Calculate Worker's Salary")
        print("3. Calculate Profit")
        print("4. Calculate Tips for a Shift")
        print("5. Calculate Total Tips for the Day")
        print("6. Exit")

        choice = input('Enter your choice (1-6): ')
        if choice == '1':
            daily_sales()  #excute function to Calculate Total Sales for the Day
        elif choice == '2':
            worker_salary()   #excute function to Calculate Worker's Salary
        elif choice == '3':
            profit()           #excute function to Calculate Profit
        elif choice == '4':
            shift_tips()      #excute function to Calculate Tips for a Shift
        elif choice == '5':
            daily_tips()      #excute function to Calculate Total Tips for the Day
        elif choice == '6':   # if user chooses "6", exit the program
            break
        else:    #error handling to ensure user inputs only numbers between 1-6
            print('Invalid choice. Please enter a number between 1 and 6.')
            
```
